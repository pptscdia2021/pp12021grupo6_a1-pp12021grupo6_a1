```python
# 1er Ejercicio: Obtener los datos completos de los empleados
# Colocar Password (en bloque python) correspondiente antes de ejecutar

#MySQL
SELECT * FROM employees;

#Desde Python
import pymysql
try:
    conexion = pymysql.connect(host='localhost',
                             user='root',
                             password='********',
                             db='employees')
    try:
        with conexion.cursor() as cursor:
            # En este caso no necesitamos limpiar ningún dato
            # Para probar, agregar al final de la sentencia SQL, por ejemplo, LIMIT 100
            cursor.execute("SELECT * FROM employees;")
 
            # Con fetchall traemos todas las filas
            employees = cursor.fetchall()
 
            # Recorrer e imprimir
            for emp in employees:
                print(emp)
    finally:
        # cierra la conexión
        conexion.close()
    
except (pymysql.err.OperationalError, pymysql.err.InternalError) as e:
    print("Ocurrió un error al conectar: ", e)
    
```

```python
# 4to Ejercicio: Obtener los empleados que son o han sido "staff". Recordar que es una consulta multi tablas.
# Colocar Password (en bloque python) correspondiente antes de ejecutar

#MySQL
SELECT emp.*
FROM employees emp
INNER JOIN titles tit
ON emp.emp_no = tit.emp_no
WHERE title= "Staff"


#Desde Python
import pymysql
try:
    conexion = pymysql.connect(host='localhost',
                             user='root',
                             password='********',
                             db='employees')
    try:
        with conexion.cursor() as cursor:
            consulta = "SELECT emp.* FROM employees emp INNER JOIN titles tit ON emp.emp_no = tit.emp_no WHERE title= 'Staff';"
            cursor.execute(consulta)
 
            # Con fetchall traemos todas las filas
            employees = cursor.fetchall()
 
            # Recorrer e imprimir
            for emp in employees:
                print(emp)
    finally:
        conexion.close()
    
except (pymysql.err.OperationalError, pymysql.err.InternalError) as e:
    print("Ocurrió un error al conectar: ", e)
```

```python
# 8vo Ejercicio: Obtener el promedio de salarios por departamento.

#MySql
# Se joinea con LEFT con la tabla salarios y departamentos para  traer siempre la info que coincide en dept_emp
SELECT dee.dept_no AS Departamento, dep.dept_name AS Nombre_Departamento, ROUND(AVG(sal.salary),2) AS Promedio_salarios
FROM dept_emp dee
LEFT JOIN salaries sal
ON sal.emp_no = dee.emp_no
LEFT JOIN departments dep
ON dep.dept_no=dee.dept_no
GROUP BY dep.dept_name


```
